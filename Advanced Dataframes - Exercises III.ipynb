```python
# 5. Join the employees and titles DataFrames together.
```

```python
url = env.get_db_url('employees')
```

```python
employees = pd.read_sql('select * from employees;', url)
```

```python
titles = pd.read_sql('select * from titles;', url)
```

```python
emps_titles_df = pd.merge(employees, titles, how='outer', 
                          on='emp_no')
emps_titles_df
```

```python
# 6. For each title, find the hire date of the employee that was 
# hired most recently with that title.
```

```python
emps_titles_df.dtypes
```

```python
emps_titles_df['hire_date'] = pd.to_datetime(emps_titles_df[
    'hire_date'])
```

```python
pd.pivot_table(data=emps_titles_df, index='title', 
               values='hire_date', aggfunc='max')
```

```python
# Write the code necessary to create a cross tabulation of the 
# number of titles by department. (Hint: this will involve a 
# combination of SQL code to pull the necessary data and 
# python/pandas code to perform the manipulations.)
```

```python
dept_emp = pd.read_sql('select * from dept_emp;', url)
```

```python
departments = pd.read_sql('select * from departments;', url)
```

```python
titles = pd.read_sql('select * from titles;', url)
```

```python
emp_titles = pd.merge(titles, dept_emp, how='outer', on='emp_no')
```

```python
emp_titles
```

```python
titles_depts = pd.merge(emp_titles, departments, how='outer', on='dept_no')
```

```python
titles_depts
```

```python
titles_depts = titles_depts[['title', 'dept_name']] 
```

```python
titles_depts
```

```python
pd.crosstab(titles_depts.title, titles_depts.dept_name)
```

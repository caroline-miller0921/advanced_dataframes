```python
from pydataset import data
import pandas as pd
import numpy as np
```

```python
# 1. Copy the users and roles DataFrames from the examples above.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
# 2. What is the result of using a right join on the DataFrames?
```

```python
pd.merge(roles, users, how='right', on='id')
```

```python
# 3. What is the result of using an outer join on the DataFrames?
```

```python
pd.merge(roles, users, how='outer', on='id')
```

```python
# 12. Create a column named mileage_difference like you did 
# in the DataFrames exercises; this column should contain 
# the difference between highway and city mileage for each 
# car.
```

```python
mpg['mileage_difference'] = mpg.hwy - mpg.cty
```

```python
mpg
```

```python
# 13. Create a column named average_mileage like you did in 
# the DataFrames exercises; this is the mean of the city 
# and highway mileage.
```

```python
mpg['average_mileage'] = (mpg.cty + mpg.hwy) / 2
```

```python
mpg
```

```python
# 14. Create a new column on the mpg dataset named 
# is_automatic that holds boolean values denoting whether 
# the car has an automatic transmission.
```

```python
mpg['is_automatic'] = mpg.trans.str.contains('auto')
```

```python
mpg
```

```python
# 15. Using the mpg dataset, find out which which manufacturer
# has the best miles per gallon on average?
```

```python
mpg.groupby('manufacturer').mean().average_mileage.max()
```

```python
# 16. Do automatic or manual cars have better miles per gallon?
```

```python
mpg.groupby('is_automatic').mean().average_mileage
```
